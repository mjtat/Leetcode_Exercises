## Two Sum I

Given an array of integers, return indices of the two numbers such that they add up to a specific target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

Example:
Given nums = [2, 7, 11, 15], target = 9,

Because nums[0] + nums[1] = 2 + 7 = 9,
return [0, 1].

**Notes**
* For a given number, detremine if it's complement is in it's array. If it is, return their indicies. Or in other words, for each element x, find a number that equals target minus x.

```python
array = [2,7,11,15]
def twoSum_I(array, target):
    array2 = array[1:]
    for num in array:
        comp = target - num
        for num2 in array2:
            if num2 == comp:
                index_1 = array.index(num)
                index_2 = array.index(num2)
                return [index_1, index_2]
    
    
   
```

```python
def binarySearch(alist, item):
    # If list length zero, return false.
    if len(alist) == 0:
        return False
    # If list length > 0 continue
    else:
        # Create midpoint
        midpoint = len(alist)//2
        # If midpoint == target, return True
        if alist[midpoint]==item:
            return True
        
        # If midpoint is not the target, continue
        else:
            # If target is less than the midpoint of the list
            if item<alist[midpoint]:
                # Recursively call binary search on the right side of the list till midpoint is found
                return binarySearch(alist[:midpoint],item)
                # Recursively call binary search on left side of the list till midpoint is found.
            else:
                return binarySearch(alist[midpoint+1:],item)
```

## Valid Palindrome

Question:
Given a string, determine if it is a palindrome, considering only alphanumeric characters
and ignoring cases.
For example,
"A man, a plan, a canal: Panama" is a palindrome.
"race a car" is not a palindrome.

```python
string = "A man, a plan, a canal: Panama"
```

```python
def isPalindrome(string):
    string = string.lower()
    formatted_string = ""
    for i in string:
        if i.isalpha() == True:
            formatted_string += i
    rev_string = ""
    j = -1
    for i in formatted_string:
        rev_string += formatted_string[j]
        j -= 1
        
    if formatted_string == rev_string:
        return True
    else:
        return False
```

```python
isPalindrome(string)
```

## Perfect Even Squares

For a a given number, determine it's divisors, then return the fraction of perfect even squares from those divisors.

For example the number 20:

2: Not a perfect even square
4: A perfect even square (4 * 4)
5: Not a perfect even square
10: Not a perfect even square
20: Not a perfect even square

Output: = 1/5

**Notes**
* Only need to check numbers up to the length of the number (bigger numbers won't divide evenly)
* Only need to check even numbers because we are looking for perfect even squares

```python
def perfectEven(number):
    valid_divisors = 0
    even_squares = 0
    for i in list(range(2, number)):
        if number % i == 0:
            valid_divisors += 1
            if i % 2 == 0 and (i ** .5) % 2 == 0:
                print("%i is a perfect even square" % i)
                even_squares += 1
            else:
                print("%i is not a perfect even square" % i)

    return even_squares / valid_divisors
```

### Given a sorted integer array where the range of elements are [0, 99] inclusive, return its missing ranges.

For example, given [0, 1, 3, 50, 75], return [“2”, “4->49”, “51->74”, “76->99”]

## O(n) solution
* Create array of full numbers
* If the difference between the ith item and the jth item is == 2, return the value in between.
* If the difference between the ith item and the jth item is not == 1, demarcate i and j.


```python
def missingRange(array):
    full_range = range(array[0], array[-1])
    missing_ranges = []
    
    for i in range(len(array)):
        if i == len(array) - 1:
            break
        
        elif array[i+1] - array[i] == 2:
            missing_no = array[i+1] - 1
            missing_ranges.append(str(missing_no))
        
        elif array[i+1] - array[i] > 2:
            missing_no = "%i->%i" %(array[i]+1, array[i+1] - 1)
            missing_ranges.append(missing_no)
            
    return(missing_ranges)
```

```python
array = [1,9,12,14,15,25]
missingRange(array)
```

### Reverse digits of an integer. 

For example: x = 123, return 321. x = -123 returns -321

```python
def reverseDigits(digits):
    string = str(digits)
    empty = ""
    j = -1
    for i in string:
        
        empty += string[j]
        j -= 1
        
    if "-" in empty:
        empty = empty[0:len(empty)-1]
        empty = -1 * float(empty)
        
    else: 
        empty = float(empty)
        
    return empty
```

```python
digits = -456
reverseDigits(digits)
```

### Question: Given a number represented as an array of digits, plus one to the number. 
* [1, 0, 1] -> [1, 0, 2]
* [1,9,9] -> [2,0,0]       

```python
def digitArray(digitarray):
    string = ''.join([str(i) for i in digitarray])
    integer = int(string) + 1 
    string = str(integer)
    new = [int(i) for i in string]
    return new
        
```

### Question:

Determine whether an integer is a palindrome. Do this without extra space.

1331 is a palindrome. 1332 is not.
256652 is a palindrome. 256642 is not.

```python
def numberPalindrome(number):
    number = str(number)
    mid = len(number) // 2
    index_1 = 0
    index_2 = -1
    for i in range(mid):
        if number[index_1] != number[index_2]:
            return False
        else:
            if number[index_1] == number[index_2]:
                index_1 += 1
                index_2 -= 1
    return True
        
```
